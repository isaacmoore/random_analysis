```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ggplot import *
import warnings
warnings.simplefilter(action = "ignore", category = FutureWarning) 
%matplotlib inline
```

```python
crime = pd.read_csv('LAPD_Crime_and_Collision_Raw_Data_for_2015.csv')
print crime.shape
crime.head()
```

```python
# Renaming the crime description column
crime.columns = [u'Date Rptd', u'DR. NO', u'DATE OCC', u'TIME OCC', u'AREA',
       u'AREA NAME', u'RD', u'Crm Cd', u'crime_desc', u'Status',
       u'Status Desc', u'LOCATION', u'Cross Street', u'Location 1']
```

```python
# verifying the column has been changed
crime.columns
```

```python
# Encoding rape
rape = list()
for i in crime.crime_desc:
    if type(i) == float:
        rape.append(0)
    elif 'RAPE' in i:
        rape.append(1)
    else:
        rape.append(0)
```

```python
# confirming the lenghts match
print len(rape)
print len(crime)
```

```python
# Assigning the encoding back to the dataframe
crime['rape'] = rape
```

```python
# confirming the encoding has been added to the dataframe.
crime.head()
```

**Checking the number of occurences in a month**

```python
occurences_per_month = {}
for i in rape_df['date'].dt.month:
    if i in occurences_per_month.keys():
        occurences_per_month[i] += 1
    else:
        occurences_per_month[i] = 1
print occurences_per_month
```

```python
month = pd.DataFrame.from_dict(occurences_per_month, orient='index')
month.index = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
month.plot(kind='line', figsize=(12,8), title='Occurences of rapes per month', marker='.', markersize=10, ylim=(0, 130))
```

**Checking the number of occurences per day in the month**

```python
occurences_by_day = {}
for i in rape_df['date'].dt.day:
    if i in occurences_by_day.keys():
        occurences_by_day[i] += 1
    else:
        occurences_by_day[i] = 1
print occurences_by_day
```

```python
pd.DataFrame.from_dict(occurences_by_day, orient='index').plot(kind='line', figsize=(12,8), 
                                                               title='Occurences of rapes by day in the motnh',
                                                              marker='.', markersize=10)
```

**Checking the number of occurences per weekday**

```python
occurences_by_weekday = {}
for i in rape_df['date'].dt.weekday:
    if i in occurences_by_weekday.keys():
        occurences_by_weekday[i] += 1
    else:
        occurences_by_weekday[i] = 1
print occurences_by_weekday
```

```python
week = pd.DataFrame.from_dict(occurences_by_weekday, orient='index')
week.index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
week.plot(kind='line', figsize=(12,8), title='Occurences of rapes by day in the week', marker='.', markersize=10)
```

## Look into mapping longitude and latitude -- Heatmap

```python
# Renaming the longtude and latitude column
rape_df.columns = [   u'date',       u'DR. NO',     u'DATE OCC',     u'TIME OCC',
               u'AREA',    u'AREA NAME',           u'RD',       u'Crm Cd',
         u'crime_desc',       u'Status',  u'Status Desc',     u'LOCATION',
       u'Cross Street',   u'lonlat',         u'rape']
```

```python
rape_df.head(3)
```

```python
def parenstrip(x):
    return x[1:-1]

# for i in rape_df.lonlat[:3]:
#     print i[1:-1]
```

```python
rape_df['lonlat'] = rape_df.lonlat.map(parenstrip)
```

```python
rape_df.head()
```

```python
lon = []
lat = []
for i in rape_df["lonlat"]:
    lon.append(str(i).split(",")[0])
    lat.append(str(i).split(",")[1])

rape_df['lon'] = lon
rape_df['lat'] = lat
rape_df.head()
```

```python
# Making a lists of lists for folium HeatMap
map_data = []
for i in rape_df['lonlat']:
    tmp = []
    # Longitude
    tmp.append(float(i.split(",")[0]))
    # Latitude
    tmp.append(float(i.split(",")[1]))
    # Occurence of rape
    tmp.append(1)
    # Append the tmp list to the master list
    map_data.append(tmp)
map_data[:10]
```

```python
import folium
from folium.plugins import HeatMap

# Focus map on Los Angeles
m = folium.Map([34.038971, -118.274169], tiles='stamentoner', zoom_start=10)

# Add Markers to the map.
# for i in rape_df['lonlat'].head(100):
#     folium.Marker([float(i.split(",")[0]) , float(i.split(",")[1])], popup="occurence of rape").add_to(m)

# Heatmap
HeatMap(map_data).add_to(m)

m.save('rape_in_LA_Heatmap.html')

m
```
